**Customer demographic Analysis**

average age of customers in each city

```python
avg_age_by_each_city = df1.groupby('city')['age'].mean()
avg_age_by_each_city
```

```python
detailed_stats = df1.groupby('city')['age'].agg(['count',    # number of customers
    'mean',     # average age
    'median',   # median age
    'std',      # standard deviation
    'min',      # minimum age
    'max'       # maximum age
]).round(2)
detailed_stats
```

```python
top_3_basic = df1(['city'].value_counts().head(3)
top_3_basic
```

```python
merged_data = pd.merge(df1,df2, on='customer_id', how='inner')
merged_data
```

top 3 cities by total number of orders

```python
top_cities = merged_data.groupby('city')['order_id'].count().sort_values(ascending=False).head(3)
top_cities
```

For each month of the year, calculate the total revenue.

```python
# Ensure order_date is datetime
df2['order_date'] = pd.to_datetime(df2['order_date'])
df2['order_date']
```

```python
# Extract year-month
df2['year_month'] = df2['order_date'].dt.to_period('M')
df2['year_month']
```

```python
# Group by year_month and sum revenue
monthly_revenue = df2.groupby('year_month')['total_price'].sum()
monthly_revenue
```

average total_price of orders for customers older than 40. 

```python
# Merge customers with orders
merged_data = pd.merge(df1, df2, on='customer_id', how='inner')
```

```python
# Filter customers older than 40
older_customers_orders = merged_data[merged_data['age'] > 40]
older_customers_orders
```

```python
# average total_price
avg_price_over_40 = older_customers_orders['total_price'].mean()
avg_price_over_40
```
